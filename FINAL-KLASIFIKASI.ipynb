```python
import pandas as pd

train = pd.read_csv("DATASET1000_METPEN.csv")
test  = pd.read_csv("DATAUJI100_METPEN.csv")

print("=== INFO DATASET ===")
print("Jumlah data latih :", train.shape)
print("Jumlah data uji   :", test.shape)

print("\n DATA LATIH")
display(train.head(10))
```

```python
print("\nDATA UJI")
display(test.head(10))
```

```python
label_map = {
    "negatif": 1,
    "positif": 0
}

train["label"] = train["emosi"].map(label_map)

print("=== CONTOH LABEL SETELAH MAPPING ===")
display(train[["teks","emosi","label"]].head(10))
```

```python
import re
import string

def clean_text(text):
    text = text.lower()
    text = re.sub(r"http\S+", "", text)
    text = re.sub(r"@\w+", "", text)
    text = re.sub(r"#\w+", "", text)
    text = re.sub(r"[0-9]", "", text)
    text = text.translate(str.maketrans("", "", string.punctuation))
    return text.strip()

train["clean_text"] = train["teks"].astype(str).apply(clean_text)
test["clean_text"]  = test["teks"].astype(str).apply(clean_text)

print("=== CONTOH HASIL PREPROCESSING ===")
display(train[["teks","clean_text"]].head())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(max_features=5000)

X_train = vectorizer.fit_transform(train["clean_text"])
X_test  = vectorizer.transform(test["clean_text"])

y_train = train["label"]

print("=== UKURAN MATRIX TF-IDF ===")
print("X_train shape :", X_train.shape)
print("X_test shape  :", X_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn_euclidean = KNeighborsClassifier(n_neighbors=7, metric='euclidean')
knn_euclidean.fit(X_train, y_train)

knn_euclid_pred = knn_euclidean.predict(X_test)

print("=== HASIL PREDIKSI KNN EUCLIDEAN (10 DATA PERTAMA) ===")
print(knn_euclid_pred)
```

```python
knn_manhattan = KNeighborsClassifier(n_neighbors=7, metric='manhattan')
knn_manhattan.fit(X_train, y_train)

knn_manhattan_pred = knn_manhattan.predict(X_test)

print("=== HASIL PREDIKSI KNN MANHATTAN (10 DATA PERTAMA) ===")
print(knn_manhattan_pred)
```

```python
hasil_knn_euclid = test.copy()
hasil_knn_euclid["prediksi_emosi"] = knn_euclid_pred

hasil_knn_manhattan = test.copy()
hasil_knn_manhattan["prediksi_emosi"] = knn_manhattan_pred

print("=== TABEL HASIL KNN EUCLIDEAN ===")
display(hasil_knn_euclid)

print("=== TABEL HASIL KNN MANHATTAN ===")
display(hasil_knn_manhattan)
```
